```python
from ipywidgets import FileUpload, Button, HBox, Layout
from IPython.core.display import HTML
from IPython.display import Javascript
new = 0

bou = Button(
    description='Nouveau',
    disabled=False,
    button_style='', # 'success', 'info', 'warning', 'danger' or ''
    tooltip='Nouveau',
    icon='' # (FontAwesome names without the `fa-` prefix)
)

f = FileUpload(
    description = "Charger",
    accept='.ipynb',  # Accepted file extension e.g. '.txt', '.pdf', 'image/*', 'image/*,.pdf'
    multiple=False  # True to accept multiple files upload else False
)


bou2 = Button(
    description='Terminal',
    disabled=False,
    button_style='', # 'success', 'info', 'warning', 'danger' or ''
    tooltip='Terminal',
    icon='' # (FontAwesome names without the `fa-` prefix)
)

def on_use(b):
    with open(next(iter(f.value)), "w+b") as i:
        i.write(f.data[0])
    url = "../notebooks/"+ next(iter(f.value))
    display(Javascript('window.open("{url}");'.format(url=url)))
    #display(HTML('<a href="../notebooks/note.ipynb">go !</a>'))

def on_click2(b):
    display(Javascript('window.open("../terminals/1");'))
    #display(HTML('<a href="../terminals/1">go !</a>'))
    
def on_click(b):
    global new
    new += 1
    i = open("new.ipynb","r")
    j = open("new{}.ipynb".format(new),"w")
    j.write(i.read())
    i.close()
    j.close()
    url = "../notebooks/new{}.ipynb".format(new)
    display(Javascript('window.open("{url}");'.format(url=url)))
    # display(HTML('<a href="../notebooks/new.ipynb">go !</a>'))
    
bou.on_click(on_click)
bou2.on_click(on_click2)
f.observe(on_use,"data")
display(HTML("""<h1 style="text-align:center; width:100%;"> Environnement JUPYTER </h1>"""))
display(HBox([bou,f,bou2],layout=Layout(width="100%",justify_content='center')))
```
