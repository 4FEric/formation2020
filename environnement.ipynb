# Environnement JUPYTER

```python
from ipywidgets import FileUpload, Button, HBox
from IPython.core.display import HTML

bou = Button(
    description='Nouveau',
    disabled=False,
    button_style='', # 'success', 'info', 'warning', 'danger' or ''
    tooltip='Nouveau',
    icon='check' # (FontAwesome names without the `fa-` prefix)
)

f = FileUpload(
    description = "Charger ancien",
    accept='.ipynb',  # Accepted file extension e.g. '.txt', '.pdf', 'image/*', 'image/*,.pdf'
    multiple=False  # True to accept multiple files upload else False
)

def on_use(b):
    with open("note.ipynb", "w+b") as i:
        i.write(f.data[0])
    display(HTML('<a href="../notebooks/note.ipynb">go !</a>'))
    
def on_click(b):
    display(HTML('<a href="../notebooks/new.ipynb">go !</a>'))
    
bou.on_click(on_click)
f.observe(on_use,"data")
display(HBox([bou,f]))
```
